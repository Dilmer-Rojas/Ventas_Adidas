```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Datos normales
normal = np.random.normal(loc=50, scale=10, size=1000)

# Sesgada positiva
sesgo_derecha = np.random.exponential(scale=10, size=1000)

# Con outliers
outliers = np.concatenate((np.random.normal(50, 5, 950), np.random.normal(100, 1, 50)))

# Graficar
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
sns.histplot(normal, kde=True, ax=axs[0], color="skyblue").set(title="Distribucin Normal")
sns.histplot(sesgo_derecha, kde=True, ax=axs[1], color="orange").set(title="Sesgo Positivo")
sns.histplot(outliers, kde=True, ax=axs[2], color="salmon").set(title="Con Outliers")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos normales
normal = np.random.normal(loc=50, scale=10, size=1000)

# Histograma
plt.figure(figsize=(8, 5))
sns.histplot(normal, kde=True, color="skyblue")
plt.title("Distribucin Normal")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()

# Generar datos sesgados a la derecha
sesgo_derecha = np.random.exponential(scale=10, size=1000)

# Histograma
plt.figure(figsize=(8, 5))
sns.histplot(sesgo_derecha, kde=True, color="orange")
plt.title("Distribucin Sesgada a la Derecha (Positiva)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()

# Invertir una distribucin exponencial para simular sesgo a la izquierda
sesgo_izquierda = -np.random.exponential(scale=10, size=1000) + 100

# Histograma
plt.figure(figsize=(8, 5))
sns.histplot(sesgo_izquierda, kde=True, color="mediumseagreen")
plt.title("Distribucin Sesgada a la Izquierda (Negativa)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()

# Mezcla de datos normales con outliers
datos_con_outliers = np.concatenate([
    np.random.normal(50, 5, 950),   # valores normales
    np.random.normal(100, 1, 50)    # outliers
])

# Histograma
plt.figure(figsize=(8, 5))
sns.histplot(datos_con_outliers, kde=True, color="crimson")
plt.title("Distribucin con Outliers")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
